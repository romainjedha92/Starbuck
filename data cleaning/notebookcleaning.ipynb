```python
import pandas as pd
df=pd.read_csv('directory.csv' , sep=",")
```

```python
df.dtypes
```

```python
df['Country'].unique()
```

```python
mask=df["Country"]=="US"
dfusa=df.loc[mask, :]
```

```python
dfusa
```

```python
dfusa['City1'] = dfusa['City'].apply(lambda x: x.title())
```

```python
dfusa
```

```python
dfusa["City_Clean"] = dfusa['City1'].astype(str) +","+" " + dfusa["State/Province"]
```

```python
dfusa
```

```python
dfusa.drop(['City1'], axis=1, inplace=True)
```

```python
dfusa.drop(['City'], axis=1, inplace=True)
```

```python
dfusa['City'] = dfusa['City_Clean']
```

```python
dfusa.drop(['State/Province'], axis=1, inplace=True)
```

```python
dfusa
```

```python
result = dfusa.groupby("City_Clean")["Store Number"].count().sort_values(ascending=False)
print(result)
```

```python
dfusastores = dfusa.groupby("City_Clean").agg(
    Count=("City", "count"),
    City=("City","first"),
    Country=("Country","first"),
    Longitude=("Longitude","first"),
    Latitude=("Latitude","first")
    

)
```

```python
dfusastores
```

```python
new = dfusastores["City"].str.split(",", n = 1, expand = True)
dfusastores["City"]= new[0]
dfusastores["Country"]= new[1]
```

```python
new
```

```python
dfusastores['City'] = dfusastores['City'].str.split(',').str[-2]
```

```python
dfusastores
```

```python
merged_df = pd.merge(dfusastores, new, on='City_Clean')
```

```python
merged_df
```

```python
merged_df.drop(['City'], axis=1, inplace=True)
```

```python
merged_df.drop(['Country'], axis=1, inplace=True)
```

```python
merged_df['City'] = merged_df[0]
```

```python
merged_df['State/Province'] = merged_df[1]
```

```python
merged_df.drop([0], axis=1, inplace=True)
```

```python
merged_df.drop([1], axis=1, inplace=True)
```

```python
merged_df
```

```python
new_order = ['Count', 'City', 'State/Province', 'Longitude', 'Latitude']

df = merged_df.reindex(columns=new_order)
```

```python
df
```

```python
df.to_csv("count_usa_store.csv", index=False)
```
