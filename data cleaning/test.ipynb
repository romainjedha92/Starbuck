#### Cela crée une nouvelle colonne “City_State” dans chaque dataset qui combine le nom de la ville et de l’état/province en une seule chaîne de caractères. 
#### Ensuite, il utilise la fonction match_cities() pour trouver la meilleure correspondance entre les noms des villes 
#### dans les deux datasets en utilisant la méthode token_set_ratio() de fuzzywuzzy.

#### Enfin, il fusionne les deux datasets en utilisant l’index des correspondances trouvées.

```python
from fuzzywuzzy import fuzz

def match_cities(city1, city2):
    return fuzz.token_set_ratio(city1, city2)

dfincome['City_State'] = dfincome['City'] + ','+ ' ' + dfincome['State/Province']
df_count_usa['City_State'] = df_count_usa['City'] + ','+ ' ' + df_count_usa['State/Province']

matches = []
for index1, row1 in dfincome.iterrows():
    best_match = None
    best_ratio = 0
    for index2, row2 in df_count_usa.iterrows():
        ratio = match_cities(row1['City_State'], row2['City_State'])
        if ratio > best_ratio:
            best_ratio = ratio
            best_match = index2
    matches.append(best_match)

dfincome['Matched_Index'] = matches
df_merged = pd.merge(dfincome, df_count_usa, left_on='Matched_Index', right_index=True)
```

Le code fourni implémente une fonction appelée `fuzzy_merge` qui effectue une fusion floue (fuzzy merge) entre deux DataFrames `df_merged` et `df2021` sur la base d'une correspondance approximative des valeurs des colonnes spécifiées.

Voici une explication détaillée du fonctionnement du code :

1. La fonction `fuzzy_merge` prend en paramètres les DataFrames `df_merged` et `df2021`, ainsi que les noms des colonnes `City_x` et `City` qui serviront de clés de fusion. Le paramètre optionnel `threshold` est utilisé pour spécifier le seuil de similarité entre les valeurs des colonnes, tandis que le paramètre `limit` définit le nombre maximum de correspondances à retourner pour chaque valeur.

2. La ligne `s = df2021[City].tolist()` extrait les valeurs de la colonne `City` du DataFrame `df2021` et les convertit en une liste.

3. La ligne `m = df_merged[City_x].apply(lambda x: process.extract(x, s, limit=limit))` applique une opération de correspondance floue en utilisant la fonction `extract` du module `process` de la bibliothèque fuzzywuzzy. Pour chaque valeur dans la colonne `City_x` du DataFrame `df_merged`, la fonction `extract` compare cette valeur aux valeurs de la liste `s` et renvoie une liste de tuples contenant les correspondances et leur score de similarité. Ces correspondances sont stockées dans la série `m`.

4. La ligne `df_merged['matches'] = m` assigne la série `m` à une nouvelle colonne `'matches'` dans le DataFrame `df_merged`.

5. La ligne `m2 = df_merged['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= threshold]))` crée une nouvelle série `m2` en parcourant la colonne `'matches'` du DataFrame `df_merged`. Pour chaque valeur de correspondance, elle vérifie si le score de similarité est supérieur ou égal au seuil spécifié. Si c'est le cas, elle extrait le premier élément du tuple correspondant (correspondance) et les concatène en une chaîne de caractères séparée par des virgules. Cette chaîne de caractères de correspondances est assignée à la colonne `'matches'` dans le DataFrame `df_merged`.

6. Finalement, la ligne `return pd.merge(df_merged, df2021, left_on=City_x, right_on=City, suffixes=('_left', '_right'))` effectue la fusion finale entre le DataFrame `df_merged` et `df2021` en utilisant les colonnes spécifiées `City_x` et `City` comme clés de fusion. Les suffixes `_left` et `_right` sont ajoutés aux colonnes correspondantes pour éviter les conflits de noms.

En résumé, la fonction `fuzzy_merge` utilise la correspondance floue pour trouver les correspondances approximatives entre les valeurs des colonnes spécifiées dans les deux DataFrames. Les correspondances sont stockées dans une nouvelle colonne `'matches'` et la fusion finale est effectuée en utilisant ces correspondances comme clés de fusion.

```python
from fuzzywuzzy import process
```

```python

from fuzzywuzzy import process
def fuzzy_merge(df_merged, df2021, City_x, City, threshold=90, limit=2):
    s = df2021[City].tolist()
    m = df_merged[City_x].apply(lambda x: process.extract(x, s, limit=limit))    
    df_merged['matches'] = m
    m2 = df_merged['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= threshold]))
    df_merged['matches'] = m2
    
    
    return pd.merge(df_merged, df2021, left_on=City_x, right_on=City, suffixes=('_left', '_right'))

        
df_merged_2 = fuzzy_merge(df_merged, df2021, 'City_x', 'City', threshold=80)
```
