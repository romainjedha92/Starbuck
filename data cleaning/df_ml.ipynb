```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données sur les villes et le nombre de Starbucks
data = {
    'City': df_2['City_y'],  # Remplacez par les noms des villes
    'Starbucks_Count': df_2['Count']  # Remplacez par les nombres correspondants
}
df = pd.DataFrame(data)

# Trier les villes par ordre décroissant du nombre de Starbucks
top_10_cities = df.nlargest(10, 'Starbucks_Count')

# Créer le graphique à barres pour le top 10 des villes
plt.figure(figsize=(10, 6))
plt.bar(top_10_cities['City'], top_10_cities['Starbucks_Count'])
plt.title('Top 10 des villes avec le plus grand nombre de Starbucks')
plt.xlabel('Ville')
plt.ylabel('Nombre de Starbucks')
plt.xticks(rotation=45)
plt.tight_layout()

# Afficher le graphique
plt.show()
```
