```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
dataset = pd.read_csv('profile_member_clean.csv')
```

```python
dataset
```

```python
#remplacer les valeurs M et F par des Int pour faciliter le ML
dataset['gender'] = dataset['gender'].replace({'M': 1, 'F':0})
dataset
```

```python
# Arrondir les valeurs de la colonne "revenue"
dataset['income'] = np.where(dataset['income'] >= 500, np.ceil(dataset['income'] / 1000) * 1000, np.floor(dataset['income'] / 1000) * 1000)
dataset['income'] = dataset['income'].astype(int)
dataset
```

```python
dataset.to_csv('profile_clean_final_version.csv', index=False)
```

```python
#Appliquer le StandardScaler sur les valeurs numériques en utilisant la bibliothèque 
colonnes_numeriques = ['age', 'income']

#créer un objet  StandardScaler
scaler = StandardScaler()

dataset[colonnes_numeriques] = scaler.fit_transform(dataset[colonnes_numeriques])
dataset
```

```python
dataset.to_csv('profile_dataiku.csv', index=False)
```
