```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('profil_ml_member.csv')
df
```

#Process de création de 24.000 lignes avec une loi normale pour créer les 60% de consommateurs non membres de Starbucks
#j'utilise les mean, std de mon dataset de base pour créer mon nouveau dataset non membres

```python
mean_age = df['age'].mean() #54.39 
std_age = df['age'].std() #17.40
mean_income = df['income'].mean()#65,435
std_income = df['income'].std() #21,633
std_age

```

#Création du nouveau dataframe

```python
df_non_membre = pd.DataFrame()

#nouvelle colonne gender
df_non_membre['gender'] = np.random.choice(['M', 'F'], size=24000, p=[0.5, 0.5])

#nouvelle colonne age
df_non_membre['age'] = np.random.normal(mean_age, std_age, size=24000)
df_non_membre['age'] = df_non_membre['age'].astype(int).clip(18, 90)  # Arrondir et limiter la plage d'âge

# Colonne income
df_non_membre['income'] = np.random.normal(mean_income, std_income, size=24000)
df_non_membre['income'] = df_non_membre['income'].astype(int).clip(30000, 120000)  # Arrondir et limiter la plage de revenu

# Colonne member
df_non_membre['member'] = 0

df_non_membre
```

```python
# Définition de la fonction d'arrondi au millier
def arrondir_au_millier(valeur):
    centaine = valeur % 1000  # Récupérer la valeur de la centaine
    millier = valeur // 1000  # Récupérer la valeur des milliers

    if centaine >= 500:
        return (millier + 1) * 1000  # Arrondir au millier supérieur
    else:
        return millier * 1000  # Arrondir au millier inférieur

# Appliquer la fonction d'arrondi à la colonne "income"
df_non_membre['income'] = df_non_membre['income'].apply(arrondir_au_millier)
df_non_membre['income'] = df_non_membre['income'].astype(int).clip(30000, 120000)

# Afficher les premières lignes du nouveau dataset
df_non_membre
```

```python
df_member_VS_non_member = pd.concat([df, df_non_membre], ignore_index = True)
df_member_VS_non_member
```

```python
df_member_VS_non_member['income'] = df_member_VS_non_member['income'].astype(int)
df_member_VS_non_member
```
