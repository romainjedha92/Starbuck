```python
import pandas as pd
```

```python
dataset = pd.read_excel("20230621 - Average Income - By State - 2019 à 2021.xlsx")
```

```python
dataset
```

```python
start_index = 0
end_index = 8
dataset = dataset.drop(dataset.index[start_index:end_index])
dataset
```

```python
dataset['City'] = dataset['Table 2. Per Capita Personal Income, by Metropolitan Area, 2019–2021']
dataset['2019'] = dataset['Unnamed: 1']
dataset['2020'] = dataset['Unnamed: 2']
dataset['2021'] = dataset['Unnamed: 3']
dataset
```

```python
columns_to_drop = ['Table 2. Per Capita Personal Income, by Metropolitan Area, 2019–2021', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7']
dataset=dataset.drop(columns_to_drop, axis=1)
dataset
```

```python
dataset=dataset.dropna(subset=['2019', '2020', '2021'])
dataset
```

```python
dataset.reset_index(drop=True, inplace=True)
dataset
income_dataset=dataset.to_csv('Income Dataset',index=False)
income_dataset
```
