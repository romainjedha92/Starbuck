#Process de création de 24.000 lignes avec une loi normale pour créer les 60% de consommateurs non membres de Starbucks

```python
#obtenir un échantillon de persona dont l'âge est compris entre 18 et 90 ans
def generate_sample(mean, std_dev, size):
    return np.random.normal(mean, std_dev, size)


def generate_sample(size):
    min_age = 18
    max_age = 90
    age_values = np.random.normal(min_age, max_age, size)
    return age_values

#définir une tranche d'income comprise entre 30,000 et 120,000 comme sur le dataset de base
def generate_sample_income(size):
    min_income = 30000
    max_income = 120000
    income_values = np.random.normal(min_income, max_income, size)
    return income_values
```

```python
#l'échantillon est basé sur une base homme/femme 50/50
sex_categories = ['M', 'F']
sex_probabilities = [0.5, 0.5]

#nombre de persona souhaités 21920
num_non_members = 21920

sex_values = np.random.choice(sex_categories, size=num_non_members, p=sex_probabilities)
age_values = generate_sample(num_non_members)
income_values = generate_sample_income(num_non_members)
```

```python
#Générer les 21 920 clients non membres
non_member_data = pd.DataFrame({
    'age': age_values,
    'gender': sex_values,
    'income': income_values
})
```

```python
#Création du Dataset clients non membres
non_member_data.to_csv('dataset_non_member_V2.csv', index=False)
```

```python
# Charger le dataset existant des membres avec une carte de fidélité
df_profile = pd.read_csv('profil_ml_member.csv')

# Concaténer les données des membres et des acheteurs sans carte de fidélité
combined_data = pd.concat([df_profile, non_member_data], ignore_index=True)

# Enregistrer les données combinées dans un nouveau fichier CSV
combined_data.to_csv('dataiku_V2.csv', index=False)
```
